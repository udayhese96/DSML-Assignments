```python
# 5. Visualize the data using R/Python by plotting the graphs for assignment no. 1 and 2. Use
# Scatter plot, bar plot, Box plot and Histogram OR Perform the data visualization operations
# using Tableau for the given dataset. Consider suitable data set.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/heart_disease_data.csv')
data.head()
```

```python
# Setting style
plt.style.use('seaborn-whitegrid')

# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='chol', hue='sex', palette='Set1', marker='o', s=100)
plt.title('Scatter Plot of Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend(title='Sex', loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Setting style
plt.style.use('seaborn-darkgrid')

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['trestbps', 'chol', 'thalach']], linewidth=2, flierprops=dict(marker='o', markersize=8))
plt.title('Box Plot of trestbps, chol, thalach')
plt.ylabel('Value')
plt.xticks(fontsize=12)  # Adjust x-axis tick label font size
plt.yticks(fontsize=12)  # Adjust y-axis tick label font size
plt.grid(True, linestyle='--', alpha=0.5)  # Add grid lines
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', bins=20, kde=True, color='skyblue', edgecolor='black', linewidth=1.2)
plt.title('Histogram of Age', fontsize=16)  # Adjust title font size
plt.xlabel('Age', fontsize=14)  # Adjust x-axis label font size
plt.ylabel('Frequency', fontsize=14)  # Adjust y-axis label font size
plt.xticks(fontsize=12)  # Adjust x-axis tick label font size
plt.yticks(fontsize=12)  # Adjust y-axis tick label font size
plt.grid(True, linestyle='--', alpha=0.5)  # Add grid lines
plt.tight_layout()
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='sex', palette='pastel')  # Using a pastel color palette
plt.title('Bar Plot of Sex', fontsize=16)  # Adjust title font size
plt.xlabel('Sex', fontsize=14)  # Adjust x-axis label font size
plt.ylabel('Count', fontsize=14)  # Adjust y-axis label font size
plt.xticks(fontsize=12)  # Adjust x-axis tick label font size
plt.yticks(fontsize=12)  # Adjust y-axis tick label font size
plt.grid(True, linestyle='--', alpha=0.5)  # Add grid lines
plt.tight_layout()
plt.show()
```

```python
# Pie chart for target distribution
target_counts = data['target'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(target_counts,
        labels=['No Heart Disease', 'Heart Disease'],
        autopct='%1.1f%%',
        startangle=140,
        colors=['#66c2ff', '#ff9999'],  # Custom colors for the pie chart wedges
        explode=(0.1, 0),  # Explode the first slice slightly for emphasis
        shadow=True,  # Add shadow to the pie chart
        textprops={'fontsize': 12, 'color': 'black'})  # Adjust text properties
plt.title('Distribution of Heart Disease', fontsize=16)  # Adjust title font size
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
# Fill Between plot with grid lines
plt.figure(figsize=(10, 6))
plt.fill_between(data.index, data['trestbps'], alpha=0.5)
plt.xlabel('Index')
plt.ylabel('trestbps')
plt.title('Fill Between Plot of trestbps')
plt.grid(True)  # Adding grid lines
plt.show()
```

```python
# Event Plot
plt.figure(figsize=(12, 6))
plt.eventplot(data['age'], orientation='horizontal', colors='r')
plt.xlabel('Age')
plt.title('c')
plt.show()
```

```python
# Stem Plot
plt.figure(figsize=(12, 6))
plt.stem(data.index, data['chol'], linefmt='b-', markerfmt='bo', basefmt='r-')
plt.xlabel('Index')
plt.ylabel('chol')
plt.title('Stem Plot of chol')
plt.show()
```
