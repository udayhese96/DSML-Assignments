```python
# Reading Data
import pandas as pd
csvFile = pd.read_csv('/content/Salary_dataset.csv')
csvFile
```

```python
csvFile.head()
```

```python
# indexing & selecting
csvFile.iloc[2:4,1:]
```

```python
import matplotlib.pyplot as plt
csvFile.describe()
```

```python
plt.figure(figsize=(12, 8))
plt.hist(csvFile['YearsExperience'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Years of Experience', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.hist(csvFile['Salary'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Salary', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Years of Experience', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
